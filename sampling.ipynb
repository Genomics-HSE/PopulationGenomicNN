```python
import numpy as np

LAMBDA_EXP = 10
```

```python
t = sorted(np.random.exponential(LAMBDA_EXP, size=100))
```

```python
max_t = t[-1]
min_t = t[0]
N = 10
a = (-np.log(max_t) + N*np.log(min_t))/(N-1)
B = (-np.log(min_t) + np.log(max_t))/(N-1)
```

```python
def to_T(time):
    return round((np.log(time)-a)/B)

def to_t(time):
    return np.e**(a+B*time)
```

```python
for i in range(1,N + 1):
    print(f"{i}: {to_t(i)}")
```

```python
print(f'{max_t} -> {to_T(max_t)}')
```

```python
for time in t:
    print(f'{time} -> {to_T(time)}')
```

```python
l = max_t/N

def to_T_2(time):
    return min(int(time/l) + 1, 10)
```

```python
for time in t:
    print(f'{time} -> {to_T_2(time)}')
```
